**Grouping**

```python
data = {'Nama': ['Jack', 'Joko', 'Ganis', 'Dyto'],
        'Umur' : [20, 23, 21, 12],
        'Alamat' : ['Surabaya', 'Jakarta', 'Kulon Progo', 'Wates'],
        'Pekerjaan' : ['Programmer', 'DB Admin', 'Web Developer', 'Programmer'],
        'Gaji': [2000, 3000, 6000, 5000]}
df = pd.DataFrame(data)
df
```

```python
df.groupby('Pekerjaan').groups
{'DB Admin': [1], 'Programmer': [0,3], 'Web Developer': [2]}
df_group = df.groupby('Pekerjaan')

for pekerjaan, group in df_group:
  print(pekerjaan)
  print(group)
```

```python
print(df_group.get_group('DB Admin'))
```

Aggregation

```python
import numpy as np
data = {'Nama': ['Jack', 'Joko', 'Ganis', 'Dyto'],
        'Umur' : [20, 23, 21, 12],
        'Alamat' : ['Surabaya', 'Jakarta', 'Kulon Progo', 'Wates'],
        'Pekerjaan' : ['Programmer', 'DB Admin', 'Web Developer', 'Programmer'],
        'Gaji': [2000, 3000, 6000, 5000]}
df = pd.DataFrame(data)
print('sebelum di kelompokkan')
print(df)

df_group = df.groupby('Pekerjaan').agg(np.mean)
print('setelah dikelompokkan')
df_group
```

```python
df_group = df.groupby('Pekerjaan').agg([np.mean, np.sum, np.std])
print('setelah dikelompokkan')
df_group
```

Merge

```python
import pandas as pd
kiri = pd.DataFrame({'Id':[2,3,1],
                     'Nama' : ['Andi', 'Joko', 'Budi']})
kanan = pd.DataFrame({'Id':[1,2,3],
                     'Nama' : ['Mike', 'Ganis', 'Jane']})
gabung = pd.merge(kiri, kanan, on='Id', sort=True)
gabung
```
